# Communication Through Noisy Channels

```python
def BinarySymmetricChanel(bitstring):
    
    result = ""
    
    for ch in bitstring:
        if random.random() < 0.1: #with prob 0.1
            if ch == "0":
                result = result + "1" #flip bit
            else:
                result = result + "0" #flip bit   
        else:
            result = result + ch
    
    return result
```

```python
a = "010101010101111000010010101001"
b = BinarySymmetricChanel(a)
print(b)
```

```python
a = ""
for i in range(5000):
    a = a + str(random.randint(0,1))
print(a)
```

```python
b = BinarySymmetricChanel(a)
print(b)
```

```python
#compare a and b
a_list = []
b_list = []

for ch in a:
    a_list.append(int(ch))
for ch in b:
    b_list.append(int(ch))
    
erroneous = 0
matches = np.logical_xor(a_list, b_list)
for i in matches:
    if i:
        erroneous += 1
print(erroneous/5000) #should be around 10%
```

```python
def calc_erroneous(a,b):
    a_list = []
    b_list = []
    
    for ch in a:
        a_list.append(int(ch))
    for ch in b:
        b_list.append(int(ch))

    erroneous = 0
    matches = np.logical_xor(a_list, b_list)
    for i in matches:
        if i:
            erroneous +=1
    return erroneous/5000
```

```python
def decode(bitstring, step):
    decoded = ""
    for i in range(0, len(bitstring)-2, step):
        if sum([int(x) for x in bitstring[i:i+step]]) > (step//2): #bit is a 1
            decoded = decoded + "1"
        else:
            decoded = decoded + "0"
    return decoded
```

```python
c = decode(a,3)
d = decode(b,3)
print(calc_erroneous(c,d))
```

```python
def com_rate(size, n):
    return math.log(size,2)/n
```

```python
print(com_rate(3,15000))
print(com_rate(5,25000))
print(com_rate(11,55000))
print(com_rate(23,115000))
```

As we can see, increasing the number of repetitions decreases the rate of communication (speed reduces). But as we can also see from above, the probability of errors goes down as well.
