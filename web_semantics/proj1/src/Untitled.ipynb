```python
osm_file = open("osm_data_1.osm", "r")
osm_soup = BeautifulSoup(osm_file, "xml")
```

```python
nodes = osm_soup.find_all("node")
for i in range(0, len(nodes)):
    if nodes[i].find("tag") == "aerialway":
        print(nodes[i], "yabai")
```

```python
#the code snippet to get aerialway nodes
aerialway_nodes = []
for node in osm_soup.find_all("node"):
    if node.find_all(k="aerialway"):
        aerialway_nodes.append(node)
print(aerialway_nodes)
#NOTE: seems like these all contain tag k="name", so we should maybe isolate these and hash them into a dicionary?
```

```python
#this finds each node that has a name and hashes its data into a dictionary
nodes = {}
for node in osm_soup.find_all("node"):
    if node.tag != None:
        if node.find(k="name") != None:
            name = node.find(k="name").get("v")
            nodes[name] = node
print(nodes)
```
