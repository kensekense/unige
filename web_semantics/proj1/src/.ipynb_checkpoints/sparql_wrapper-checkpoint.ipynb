```python
from SPARQLWrapper import SPARQLWrapper
from bs4 import BeautifulSoup
import requests
```

```python
repository_link = "http://192.168.1.103:7200/repositories/WEB2019"
query = """
PREFIX :<http://www.topografix.com/GPX/1/1>
SELECT ?c ?t
WHERE {{?c :hassurround :nd353086852.}}
"""
db = SPARQLWrapper(repository_link)
db.setQuery(query)
db.method = "POST"
db.setReturnFormat('json')
db.queryType = "SELECT"
result = db.query()
fp = result.convert()
```

```python
#sample query 
query_results = []
for i in range(0, len(fp['results']['bindings'][1:])):
    query_results.append(fp['results']['bindings'][1:][i]['c']['value'])
print(query_results)
```

```python
def query_RDF (query):
    '''
    take a string as query, and queries our database and returns the relevant queries as a list
    '''
    repository_link = "http://192.168.1.103:7200/repositories/WEB2019"
    db = SPARQLWrapper(repository_link)
    db.setQuery(query)
    db.method = "POST"
    db.setReturnFormat('json')
    db.queryType = "SELECT"
    result = db.query()
    fp = result.convert()
    
    query_results = []
    for i in range(0, len(fp['results']['bindings'][1:])):
        query_results.append(fp['results']['bindings'][1:][i]['c']['value'])
    return query_results
```

```python
a = query_RDF("""PREFIX :<http://www.topografix.com/GPX/1/1>
SELECT ?c ?t
WHERE {{?c :hassurround :nd353086852.}}""")
print(a)
```

```python
query = """
PREFIX dataid: <http://dataid.dbpedia.org/ns/core#>
PREFIX dct: <http://purl.org/dc/terms/>
PREFIX dcat:  <http://www.w3.org/ns/dcat#>

SELECT distinct ?file ?latestVersion ?mediatype WHERE {
?dataset dataid:artifact <https://databus.dbpedia.org/denis/ontology/dbo-snapshots> .
?dataset dcat:distribution ?distribution .
    ?distribution dcat:downloadURL ?file ;
     dct:hasVersion ?latestVersion ;
      # see all available mediatypes with 
     # dcat:mediaType ?mediaType .
                  dcat:mediaType <http://dataid.dbpedia.org/ns/mt#TextTurtle> . 
    {
            SELECT (?version as ?latestVersion) WHERE { 
                ?dataset dataid:artifact <https://databus.dbpedia.org/denis/ontology/dbo-snapshots> . 
                ?dataset dct:hasVersion ?version . 
            } ORDER BY DESC (?version) LIMIT 1 
} 
}
"""
url = "https://databus.dbpedia.org/repo/sparql/"
db = SPARQLWrapper(url)
db.setQuery(query)
db.method = "POST"
db.setReturnFormat('json')
db.queryType = "SELECT"
result = db.query()
fp = result.convert()
```

```python
ttl_url = fp['results']['bindings'][0]['file']['value']
ttl_file = requests.get(ttl_url)
open('./onto_snap.ttl', 'wb').write(ttl_file.content)
print("downloaded ttl file onto_snap.ttl")
```

```python
def query_DBp (query):
    '''
    make querying faster
    '''
    db = SPARQLWrapper("http://dbpedia.org/sparql/")
    db.setQuery(query)
    db.method = "POST"
    db.setReturnFormat('json')
    db.queryType = "SELECT"
    result = db.query().convert()
    
    return result
```

```python
a = query_DBp("""
PREFIX geo:  <http://www.w3.org/2003/01/geo/wgs84_pos#>
SELECT ?s ?name ?value ?lat ?lng
WHERE { 
    ?s a <http://dbpedia.org/ontology/PopulatedPlace> .
    ?s <http://dbpedia.org/property/name> ?name .
    OPTIONAL { ?s <http://dbpedia.org/property/populationTotal> ?value . }
    FILTER (?lng > -8.64 AND ?lng < 2.1 AND ?lat < 61.1 AND ?lat > 49.35 )
    ?s geo:lat ?lat .
    ?s geo:long ?lng .
}
""")
```
