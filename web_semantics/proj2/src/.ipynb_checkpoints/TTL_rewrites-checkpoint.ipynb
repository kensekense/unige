```python
#REWRITE THE TTL FILE
original_ttl = open("./../data-sources/jrc-list.ttl", "r")
lines = []
for line in original_ttl:
    lines.append(line) #parse the turtle file as is line by line

#go through the lines and fix according to the schematic
#briefDescription -> description
#geographicCoverage -> geography (:tags)
#primaryEnvironmentalDomain -> domain (:tags)
#:startYear -> :startYear (int)
#:endYear -> :endYear (:unknown or int)
output_ = []
for item in lines:
    
    newLine = item
    changed = False
    split_view = item.split() #should split the line [word, word, etc...]
    
    if len(split_view) < 1: #empty line
        newLine = "\n"
        
    elif split_view[0] == ":briefDescription": #briefDescription -> description
        #need to add ":description" + theRest
        newLine = ":description {0}".format(" ".join(split_view[1:]))
        changed = True
    
    elif split_view[0] == ":geographicCoverage": #geographicCoverage -> geography
        #need to add ":geography" + theRest
        newLine = ":geography :{0}".format(" ".join(split_view[1:]).split("\"\"\"")[1])
        changed = True
    
    elif split_view[0] == ":primaryEnvironmentalDomain":
        #need to add ":domain" + theRest
        newLine = ":domain :{0}".format(split_view[1])
        changed = True
    
    elif split_view[0] == ":startYear":
        #need to add ":startYear" + theRest
        newLine = ":startYear {0}".format(" ".join(split_view[1:]).split("\"\"\"")[1]) #we want just a number here
        changed = True
    
    elif split_view[0] == ":endYear":
        #need to add ":endYear" + theRest
        if split_view[1] == "Unknown":
            newLine = ":endYear :Unknown"
            changed = True
        else:
            newLine = ":endYear {0}".format(" ".join(split_view[1:]).split("\"\"\"")[1]) #we want just a number here
            changed = True
    
    #append proper choice of newLine
    if changed: #there is a tab
        output_.append("   " + newLine)
    else:
        output_.append(newLine)

#test print
for thing in output_:
    print(thing)
```
