```python
#simulate the counts, we can also consider populating them via some other method
vocab = {}
q = {}
e = {}
tags = {'GENE', 'NOGENE'}

def viterbi (sentence):
    '''
    Input is a LIST of words, called a 'sentence'
    '''
    n = len(sentence) #define n to be the length of the sentence (in terms of words)
    trellis = {}
    trellis[(-1,"*","*")] = 1 #set prob of beginning to be 1
    bp = {} #backpointers
    
    #set a list to be the set of possible tags at position k, which is every tag at every position
    S = {}
    S[-2] = ["*"] #set S -2 and -1 to be the starting symbols 
    S[-1] = ["*"]
    for k in range(n):
        S[k] = tags #all possible tags
    
    #viterbi proper
    for k in range(n): #go for each position k through the entirety of the sentence
        
        #check if word is in our vocab, otherwise use the _RARE_ tag
        if sentence[k] in vocab:
            word = sentence[k]
        else:
            word = "_RARE"
        
        #trellis proper
        for u in S[k-1]:
            for v in S[k]:
                
                #establish we're looking for the max
                maxProb = 0
                maxClass = ""
                
                for w in S[k-2]:
                    prob = trellis.get((k-1, w, u), 0)*q.get((v,w,u),0)*e.get((word,v),0)
                    if prob > maxProb:
                        maxProb = prob
                        maxClass = w #giving the proper tag
                
                if maxProb > 0: #set the actual value to the trellis and backpointer
                    trellis[(k,u,v)] = maxProb
                    bp[(k,u,v)] = maxClass
        
        #account for STOP keyword, we want to tag the last word according to the STOP keyword
        maxProb = 0
        maxU, maxV = "", ""
        for u in S[n-2]:
            for v in S[n-1]:
                prob = trellis.get((n-1,u,v),0)*q.get(("STOP",u,v),0) #no e here?
                if prob > maxProb:
                    maxProb = prob
                    maxU = u
                    maxV = v
                    
    #tagging proper
    y = {}
    
    #tag the back first
    if maxU == "":
        y[n-2] = "X" #factoring in if the path is bad?
    else:
        y[n-2] = maxU
    if maxV == "":
        y[n-1] = "X"
    else:
        y[n-1] = maxV
    
    #tag the rest
    for k in range (n-3, -1, -1): #going backwards
        try:
            y[k] = bp[(k+2, y[k+1], y[k+2])] #will give us the tag with the best
        except:
            y[k] = "X" #factoring in if there is no path?
    
    return [y[k] for k in sorted(y)] #return the list
```
