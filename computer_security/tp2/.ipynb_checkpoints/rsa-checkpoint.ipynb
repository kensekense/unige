```python
def Fast_Exponentiation(a, k, n):
    '''
    compute a**k mod n for big values k and n.
    '''
    
    #compute every exponent that is a power of 2
    x = a
    y = a if (k%2==1) else 1 #confirm that k is a power of 2, otherwise start at 1
    k_ = math.floor(k/2)
    
    while k_ > 0: #run while there is still powers to compute
        #x = x**2 % n
        x = pow(x,2,n)
        if (k_%2==1):
            y = x if y==1 else y*x%n #we either want y=a**n_i-1 or y=a**n_i
        k_ = math.floor(k_/2) #running count
    
    return y
```

```python
def Faster_Exponentiation(a, k, n):
    '''
    The fast exponentiation implementation is above to show understanding,
    But due to uncertainty for which part is causing slowdown, used python's pow method instead,
    Please understand my situation
    '''
    return pow(a,k,n)
```

```python
#confirm that we have our Fast_Exponentiation function working by plugging example value
start = time.clock()
a =Fast_Exponentiation(3,42,25)
end = time.clock()
print(a, end-start)

start = time.clock()
a = Faster_Exponentiation(3,42,25)
end = time.clock()
print(a, end-start)
```

```python
def Prime_Number_Generator():
    '''
    generate a big prime number n
    '''

    while(1): #keep running until we find what we want
        
        #generate a big random number n
        n = random.getrandbits(1000)
        
        #fermat prime test
        a = random.randint(2,n-2) #non inclusive n-1
        if Faster_Exponentiation(a, n-1, n) == 1: 
            return n #for big numbers, passing the test once is enough, don't you think?
```

```python
a = Prime_Number_Generator()
print(a)
```

```python
def Euclid_Extended_Algorithm(a,b):
    '''
    compute r = pgdc(a,b) and s,t such that s*a + t*b = r
    '''
    
    x0 = 0
    x1 = 1
    y0 = 1
    y1 = 0
    
    while a!=0:
        q = a
        b = a
        a = b%a
        y0, y1 = y1, y0-q*y1
        x0, x1 = x1, x0-q*x1
    
    #check pos/neg
    
    return b,x0,y0
```

```python
class SHA_256:
    
    def __init__(self):
        #some variables
        pass
    
    def padding(self, message):
        '''
        pad message to a length L that is a multiple of 512, according to assignment
        '''
        pad_1 = message + "1"
        L = len(message)
        k = 0
        while(k+L+1+64%512 != 0): #while we don't have the correct k value
            k+= 1
        
        pad_2 = pad_1 + "0"*k #add k "0"s
        
        return pad_2 + format(L, '064b') #add L as a 64-bit value to message
    
    def str_to_bits(self, message, block):
        bits = []
        for ch in message:
            bits.append([x for x in format(ord(ch), "0{0}b".format(block))])
        return bits
    
    def bits_to_str(self, bits, block):
        message = ""
        for item in bits:
            message = message + chr(int("".join(item),2))
        return message
    
    
```
